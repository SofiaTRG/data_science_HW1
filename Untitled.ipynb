8. There's seems to be a problem with the calculation for team "Crotone" in league "Serie_A" in year 2017. The problem is most likely with column "pts". 

```python
# question 9
teams_fail_check = db.loc[ : , ['league', 'team', 'year', 'scored', 'conceded']]
teams_fail_check['failed'] = teams_fail_check.conceded > teams_fail_check.scored
teams_fail_check
```

```python
most_fails = teams_fail_check.loc[teams_fail_check['failed'] == True]
most_fails.replace(True, 1)
most_failed_leagues = most_fails.groupby('league')['failed'].sum()
most_failed_leagues.idxmax()
```

9. The league with most failed teams is "la_liga"
